##### Since the list results empty, every float columns has null values, then they can't be converted in the integer type.  


##### The next step, optimizing the dataframe by converts to better types

```python
# Which object type column can be numeric ones.
loans_first_5[obj_cols]
```

###### Candidates of object columns to be numeric: 
- term (taking off the string 'months')
- int_rate (taking off the string '%')
- revol_util (taking off the string '%')

###### Candidates  of object columns to be categoric
- All, because all have less than 50% of unique values. 

```python
go_to_numeric = ['term','int_rate','revol_util']
go_to_categoric = []
for col in obj_cols:
    if col not in go_to_numeric:
        go_to_categoric.append(col)
        
go_to_categoric
```

```python
# float_cols
total_memory_needs = 0

# column already optimized: id

# We are going to optimize object columns that can be 
# converted in numeric and category ones. Also we can optimize 
# the float columns (automated optimization by pandas.to_numeric)
# to pic the right amount of bits in float to every float column


loans_chunks = pd.read_csv("loans_2007.csv",chunksize=1000)
for chunk in loans_chunks: 
    
    for col in go_to_numeric: 
        if col == 'term':
            chunk[col] = chunk[col].str.replace('months','')
            chunk[col] = chunk[col].str.replace(' ','')
            chunk[col] = pd.to_numeric(chunk[col])
        elif col == 'int_rate':
            chunk[col] = chunk[col].str.replace('%','')
            chunk[col] = pd.to_numeric(chunk[col])
        elif col == 'revol_util': 
            chunk[col] = chunk[col].str.replace('%','')
            chunk[col] = pd.to_numeric(chunk[col])
    
    for col in float_cols: 
        chunk[col] = pd.to_numeric(chunk[col])
    
    for col in go_to_categoric: 
        chunk[col] = chunk[col].astype('category')
    
    total_memory_needs += chunk.memory_usage(deep=True).sum()

print(total_memory_needs/(1024*1024), "Megabytes")
        
```

##### The new total memory needs are 23.3 Mb for the full data frame.
